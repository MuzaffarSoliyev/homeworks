```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np

img=mpimg.imread('cat.bmp')
print(np.shape(img))
fr1 = np.zeros((len(img),len(img[1])))
fr2 = np.zeros((len(img),len(img[1])))
fr3 = np.zeros((len(img),len(img[1])))
#print(np.shape(fr1))
for i in range(len(img)):
    for j in range(len(img[i])):
        fr1[i][j] = img[i][j][0]
        fr2[i][j] = img[i][j][1]
        fr3[i][j] = img[i][j][2]
#print(img)
#print(fr1,fr2,fr3)
plt.imshow(fr1)
plt.show()
plt.imshow(fr2)
plt.show()
plt.imshow(fr3)
plt.show()
np.fft.fft2(fr1)
```
