```python
import numpy as np
data = np.load('problem5_data.npy')
print(data)
print(np.shape(data))
```

```python
X = []
Y = []
for i in data:
    X.append(i[0])
    Y.append(i[1])
import matplotlib.pyplot as plt
plt.plot(X,Y,'r.')
plt.show()
```

```python
points = np.linspace(np.min(X), np.max(X), 100)
print(points)
```

```python
n = 7
A = np.vander(X, n+1)
print(np.shape(A))
```

```python
from scipy.linalg import pinv
coeff = pinv(A) @ Y
print(coeff)
```

```python
polinom = np.poly1d(coeff)
print(polinom)
```

```python
y = polinom(X)
y_ = polinom(points)
plt.plot(points,y_,'k')
plt.plot(X,Y,'r.')
plt.show()
```
