```python
import numpy as np
def perpendicular(x1,y1,x2,y2,x3,y3,x):
    m_a = (x2-x1)/(y2-y1)
    return (-1)*(m_a)*(x - ((x1+x2)/2)) + (y1+y2)/2
def circle(x1,y1,x2,y2,x3,y3):
    m_a = (y2-y1)/(x2-x1)
    m_b = (y3-y2)/(x3-x2)
    x = (m_a*m_b*((y1-y3) + m_b*(x1+x2) - m_a*(x2+x3)))/(2*(m_b-m_a))
    y = perpendicular(x1,y1,x2,y2,x3,y3,x)
    r = np.sqrt((x-x1)**2 + (y-y1)**2)
    return x,y,r
#print(circle(1.,0.,0.,1.,-1,0.))
def distance(x,y,r,x1,y1):
    dis = np.sqrt((x-x1)**2 + (y-y1)**2)
    return np.abs(r-dis)
#print(distance(0.,0.,2.,10.,0.))
```
