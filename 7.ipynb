```python
import numpy as np
from sklearn.datasets import load_digits
digits = load_digits()
print (np.shape(digits.data))
print (np.shape(digits.target))
print (digits.data[0])
print (digits.target[0])
data = []
target = []
for i in range(len(digits.target)):
    #print(i)
    if int(digits.target[i])== 3 or int(digits.target[i])== 5 or int(digits.target[i])== 8:
        data.append(digits.data[i])
        target.append(digits.target[i])
print(len(data))
print(len(target))
```

```python
import matplotlib.pyplot as plt
plt.matshow(digits.images[3]) 
plt.show() 
plt.matshow(digits.images[5]) 
plt.show() 
plt.matshow(digits.images[8]) 
plt.show() 
```

```python
from sklearn.manifold import TSNE
X = data
X_embedded = TSNE(n_components=2).fit_transform(X)
```

```python
print(X_embedded)
print(target)
```

```python
#target = digits.target
target = np.array(target)
result = np.hstack((X_embedded, target.reshape((-1,1))))
print(result)
```

```python
import matplotlib.pyplot as plt
colors = {3:'r.', 5:'g.',8:'k.'}
for i in result:
    if int(i[2]) == 3 or int(i[2]) == 5 or int(i[2]) == 8:
        plt.plot(i[0],i[1],colors[int(i[2])])
plt.show()
```

```python
from sklearn.manifold import Isomap
embedding = Isomap(n_components=2)
X_transformed = embedding.fit_transform(X)
print(X_transformed)
```

```python
#target = digits.target
result = np.hstack((X_transformed, target.reshape((-1,1))))
print(result)
```

```python
import matplotlib.pyplot as plt
colors = {3:'r.', 5:'g.',8:'k.'}
for i in result:
    if int(i[2]) == 3 or int(i[2]) == 5 or int(i[2]) == 8:
        plt.plot(i[0],i[1],colors[int(i[2])])
plt.show()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_trans = pca.fit_transform(X)
```

```python
#target = digits.target
result = np.hstack((X_trans, target.reshape((-1,1))))
print(result)
```

```python
import matplotlib.pyplot as plt
colors = {3:'r.', 5:'g.',8:'k.'}
for i in result:
    if int(i[2]) == 3 or int(i[2]) == 5 or int(i[2]) == 8:
        plt.plot(i[0],i[1],colors[int(i[2])])
plt.show()
```
